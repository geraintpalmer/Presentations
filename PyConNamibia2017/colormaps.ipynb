```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
%matplotlib inline
```

```python
image = img.imread("tex/testimage.JPG")
```

```python
image = image[:,:,0]
```

```python
magic_square = [
    [47, 58, 69, 80, 1, 12, 23, 34, 45],
    [57, 68, 79, 9, 11, 22, 33, 44, 46],
    [67, 78, 8, 10, 21, 32, 43, 54, 56],
    [77, 7, 18, 20, 31, 42, 53, 55, 66],
    [6, 17, 19, 30, 41, 52, 63, 65, 76],
    [16, 27, 29, 40, 51, 62, 64, 75, 5],
    [26, 28, 39, 50, 61, 72, 74, 4, 15],
    [36, 38, 49, 60, 71, 73, 3, 14, 25],
    [37, 48, 59, 70, 81, 2, 13, 24, 35]]
```

```python
def plot_church(cmap, reverse=False):
    if reverse:
        cmap_use = cmap + "_r"
    else:
        cmap_use = cmap
    fig, ax = plt.subplots(1)
    plt.imshow(image, cmap=cmap_use)
    plt.tick_params(axis='both', which='both', bottom='off', top='off', left='off', right='off', labelbottom='off', labelleft='off')
    plt.colorbar(ticks=[])
    plt.title(cmap, fontsize=20)
    plt.tight_layout()
    return fig
```

```python
def plot_magicsquare(cmap, reverse=False):
    if reverse:
        cmap_use = cmap + "_r"
    else:
        cmap_use = cmap
    fig, ax = plt.subplots(1)
    plt.pcolor(magic_square, cmap=cmap_use)
    plt.tick_params(axis='both', which='both', bottom='off', top='off', left='off', right='off', labelbottom='off', labelleft='off')
    plt.colorbar(ticks=[])
    plt.title(cmap, fontsize=20)
#     plt.tight_layout()
    return fig
```

```python
colormaps = ['jet', 'gray', 'cool', 'hot', 'coolwarm', 'BrBG', 'Oranges', 'viridis', 'magma', 'inferno', 'flag']
to_reverse = [True, False, True, True, True, False, True, False, False, False, False]
```

```python
for cmap, rev in zip(colormaps, to_reverse):
    plot_church(cmap, reverse=rev);
```

```python
for cmap, rev in zip(colormaps, to_reverse):
    plot_magicsquare(cmap, reverse=rev);
```

```python
def f(x, y):
    return -(x**2+3*y**2)*math.exp(-x**2-y**2)
```

```python
xs = np.arange(-2, 2, 0.1)
ys = np.arange(-2, 2, 0.1)
z = []
for x in xs:
    z.append([])
    for y in ys:
        z[-1].append(f(x, y))
```

```python
plt.pcolor(z, cmap='viridis')
plt.grid(True)
```

```python
p = plot_magicsquare('plasma');
p.savefig('tex/magicsquare_plasma.pdf')
```
