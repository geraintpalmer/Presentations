```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
xs = [2]
ys = [2]
zs = [2]
```

```python
t = 0.01
```

```python
a = 10
b = 28
c = 8/3
```

```python
def lorenz(xs, ys, zs, a, b, c, t):
    x = xs[-1]
    y = ys[-1]
    z = zs[-1]
    dx = a*(y - x)
    dy = x*(b - z) - y
    dz = (x*y) - (c*z)
    xs.append(x + t*dx)
    ys.append(y + t*dy)
    zs.append(z + t*dz)
```

```python
for i in range(5000):
    lorenz(xs, ys, zs, a, b, c, t)
```

```python
cmap = matplotlib.cm.get_cmap('viridis')
```

```python
fig = plt.figure(figsize=(18, 10))
ax = fig.add_subplot(111, projection='3d')
for i in range(4999):
    ax.plot(xs[i:i+2], ys[i:i+2], zs[i:i+2], c=cmap(i/5000))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=0, vmax=t*5000))
sm._A = []
cbar = plt.colorbar(sm)
cbar.ax.set_title('Time')
fig.savefig('lorenz.pdf')
```

```python
x1s = [2]
y1s = [2]
z1s = [2]

x2s = [2.0001]
y2s = [2.0001]
z2s = [2.0001]
```

```python
for i in range(5000):
    lorenz(x1s, y1s, z1s, a, b, c, t)
    lorenz(x2s, y2s, z2s, a, b, c, t)
```

```python
diffx = [x1 - x2 for x1, x2 in zip(x1s, x2s)]
diffy = [y1 - y2 for y1, y2 in zip(y1s, y2s)]
diffz = [z1 - z2 for z1, z2 in zip(z1s, z2s)]
```

```python
fig, axarr = plt.subplots(3)
axarr[0].plot(diffx)
axarr[0].set_ylabel('$\Delta x$', rotation=0)
axarr[1].plot(diffy)
axarr[1].set_ylabel('$\Delta x$', rotation=0)
axarr[2].plot(diffz)
axarr[2].set_ylabel('$\Delta x$', rotation=0)
plt.show()
```
