```python
%matplotlib inline
import random
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
def logisticmap(x, y):
    return 4*x*(1-x), (x+y)%1
```

```python
random.seed(1)
```

```python
xs = [0.2+(0.15*(random.random())) for _ in range(150)]
ys = [0.5+(0.15*(random.random())) for _ in range(150)]
```

```python
def next_iter(xs, ys):
    newxs = []
    newys = []
    for coord in range(150):
        x, y = logisticmap(xs[coord], ys[coord])
        newxs.append(x)
        newys.append(y)
    return newxs, newys
```

```python
def plot_lm(xs, ys, ax, i, cmap):
    ax.scatter(xs, ys, s=8, c=cmap(i/9))
    ax.set_ylim(0, 1)
    ax.set_xlim(0, 1)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("Iteration " + str(i), fontsize=16)
```

```python
cmap = matplotlib.cm.get_cmap('plasma')
```

```python
fig, axarr = plt.subplots(3, 3, figsize=(8, 8))
for i in range(9):
    plot_lm(xs, ys, axarr[i//3, i%3], i, cmap)
    xs, ys = next_iter(xs, ys)
plt.tight_layout()
fig.savefig('logistic_map.pdf')
```
