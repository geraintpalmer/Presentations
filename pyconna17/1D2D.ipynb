```python
import random
```

```python
random.seed(52)
```

```python
one_dimention = [random.lognormvariate(0.4, 0.2) for _ in range(100)] + [random.random()*2.5 for _ in range(4)]
```

```python
plt.hist(one_dimention)
plt.title('plt.hist(xs)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('hist.pdf')
```

```python
plt.boxplot(one_dimention)
plt.title('plt.boxplot(xs)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('box.pdf')
```

```python
plt.violinplot(one_dimention)
plt.title('plt.violinplot(xs)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('violin.pdf')
```

```python
xs = [1, 2, 3, 4, 5]
ys = [7, 2, 3, 4, 3]
```

```python
plt.scatter(xs, ys)
plt.title('plt.scatter(xs, ys)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('scatter.pdf')
```

```python
plt.bar(xs, ys)
plt.title('plt.bar(xs, ys)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('bar.pdf')
```

```python
plt.plot(xs, ys)
plt.title('plt.plot(xs, ys)', fontsize=30, fontname='Courier New', y=1.05)
plt.savefig('line.pdf')
```
